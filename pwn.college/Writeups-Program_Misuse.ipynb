### ssh into your challenge instances
```bash
# 1. generate a ssh key
ssh-keygen -f ~/.ssh/keys/pwncollegekey -N ''   # generate files pwncollegekey and pwncollegekey.pub

# 2. connect into your challenge instance
ssh -i ~/.ssh/keys/pwncollegekey hacker@dojo.pwn.college
```

### OR 

```bash
# 2. config ssh
cat ~/.ssh/config

        Host dojo.pwn.college
        User hacker
        IdentityFile ~/.ssh/keys/pwncollegekey

# 3. ssh to host 
ssh dojo.pwn.college
```

## Offline install of [oh-my-zsh](https://github.com/ohmyzsh/ohmyzsh) & [zsh-syntax-highlighting](https://github.com/zsh-users/zsh-syntax-highlighting)

---

### Local :
```bash
# download oh-my-zsh
scp -i ~/.ssh/keys/pwncollegekey ./ohmyzsh-master.zip hacker@dojo.pwn.college:~/

# download https://gist.github.com/hewerthomn/65bb351bf950470f6c9e6aba8c0c04f1
scp -i ~/.ssh/keys/pwncollegekey ./install-oh-my-zsh.sh hacker@dojo.pwn.college:~/

# download zsh-syntax-highlighting
scp -i ~/.ssh/keys/pwncollegekey ./zsh-syntax-highlighting-master.zip hacker@dojo.pwn.college:~/
```

### server :
```bash
unzip ./master.zip  
mv ./master ./.oh-my-zsh

chmod +x install-oh-my-zsh.sh && ./install-oh-my-zsh.sh

unzip ./zsh-syntax-highlighting-master.zip
mv ./zsh-syntax-highlighting-master ./zsh-syntax-highlighting

echo "source ${(q-)PWD}/zsh-syntax-highlighting/zsh-syntax-highlighting.zsh" >> ${ZDOTDIR:-$HOME}/.zshrc
```
